## 1A: English Dictionary App :-

## This project uses the skills learnt from week 1 to week 6. The aim of this project is to create an English Dictionary app that returns definitions of English words. When the application is started, it displays a menu as follows:

Main Menu

Add a new word

Find the meaning

Update a word

Exit


```python
import json

my_dict = {}
print("Main Menu : \n 1.Add a new word \n 2.Find the meaning \n 3.Update a word \n 4.Exit\n")

while True:
    try:
        action=int(input('Please select from the given mainmenu(1,2,3,4):\n'))
    except ValueError :
        print('!!!!Please enter interger values only!!!\n')
        print('*'*50,"\n")
    else:
        if action == 4:
            e = input("Are you sure you want to Exit ? (yes/No) : ")
            if e=='Yes' or e=="yes":
                print("Thank you , Have a nice day...!")
                print('*'*50,"\n")
                break
            else:
                continue
        
        elif action == 1:
            word = input("Add a new Word : ")
            meaning = input("Meaning of the Word : ")
            my_dict[word] = meaning
            with open ("words.txt",'w') as f:
                json.dump(my_dict, f)
                
            print("Words added successfully....! \n")
            print('*'*50,"\n")
            
                
        elif action == 2 :
            find = input("Find the meaing of the Entered word : ")
            with open ("words.txt",'r') as f:
                my_dict = json.load(f)
                if find in my_dict:
                    print(f"Meaning of the word you entered is : {my_dict[find]}")
                    print('*'*50,"\n")
                    continue
                else:
                    print("Word not found")
                    print('*'*50,"\n")
                    
        elif action == 3 :
            word = input("Enter a word which you want to ubdate : ")
            with open ("words.txt",'r') as f:
                my_dict = json.load(f)
                if word in my_dict:
                    meaning = input("Enter a new word : ")
                    my_dict[word] = meaning
                    with open ("words.txt",'w') as f:
                        json.dump(my_dict,f)
                    print(f"Has been updated to {meaning}")
                    print('*'*50,"\n")
                    continue
                else:
                    print("Word not Found")
                    print('*'*50,"\n")
                    
        else:
            print('!!!Please chose the correct option from the Mainmenu!!!')
            print('*'*50,"\n")
```
